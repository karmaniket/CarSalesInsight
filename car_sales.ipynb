```python
from sklearn.linear_model import LinearRegression as lr
```

```python
import seaborn as sbn
```

```python
sbn.set()
```

```python
cars_data = pd.read_csv('cars_2000_2024.csv')
```

```python
print(cars_data.head())
```

```python
def plot_sales(car_brand):
    brand_data = cars_data[cars_data['Car brand'] == car_brand]
    x = brand_data[['Year']].values
    y = brand_data[['Number of Cars sold']].values
    model = lr()
    model.fit(x, y)
    y_pre = model.predict(x)
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', label= 'Actual Sales')
    plt.plot(x, y_pre, color= 'red', label= 'Predicted Sales', linewidth= 2)
    plt.xlabel('Year')
    plt.ylabel('Number of cars sold')
    plt.title(f'Sales of {car_brand} (2000 - 2024)')
    plt.legend()
    plt.show()
    print(f'Intercept: {model.intercept_}')
    print(f'Coefficient: {model.coef_[0]}')
```

```python
plot_sales('toyota')
```

```python
plot_sales('ford')
```
